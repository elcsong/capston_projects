### Investigate the vairables
__Findings__:
* Zero trop_distance having total_amount
* Negative value of total_amount

```python
# Sort by 'trip_distance' from maximum to minimum value
trip_distance = df.sort_values(by='trip_distance', ascending=False)
trip_distance
```

```python
# Sort by 'total_amount'
total_amount = df.sort_values(by='total_amount', ascending=False)[['total_amount']]
total_amount.head(20)
```

```python
# Sort the data by total amount and print the bottom 20 values
total_amount.tail(20)
```

```python
# How many of each payment type are represented in the data?
df.payment_type.value_counts()
```

<hr>According to the data dictionary, the payment method was encoded as follows:

1 = Credit card  
2 = Cash  
3 = No charge  
4 = Dispute  
5 = Unknown  
6 = Voided trip

```python
# What is the average tip for trips paid for with credit card?
avg_cc_tip = df[df['payment_type'] == 1].tip_amount.mean()
print(f"Avg. cc tip: {avg_cc_tip:.2f}")

# What is the average tip for trips paid for with cash?
avg_cash_tip = df[df['payment_type'] == 2].tip_amount.mean()
print(f"Avg. cash tip: {avg_cash_tip:.2f}")
```

```python
# How many times is each vendor ID represented in the data?
df['VendorID'].value_counts()
```

```python
# What is the mean total amount for each vendor?
df.groupby('VendorID').agg({'total_amount':'mean'})
```

```python
# Filter the data for credit card payments only
credit_card = df[df['payment_type']==1]

# Filter the credit-card-only data for passenger count only
credit_card['passenger_count'].value_counts()
```

```python
# Calculate the average tip amount for each passenger count (credit card payments only)
credit_card.groupby('passenger_count').agg({'tip_amount':'mean'})
```
