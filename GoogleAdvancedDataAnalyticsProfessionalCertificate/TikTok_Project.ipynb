#### Task2c. Understand the data - Investigate the variables

```python
# What are the different values for claim status and how many of each are in the data?
data.claim_status.value_counts()
```

* The count of claim status are quite balanced.

Next, examine the engagement trends associated with each different claim status.

```python
# What is the average view count of videos with "claim" status?
avg_view_count = data[data['claim_status'] == 'claim']
print(f"Avg. view count of videos with 'claim': {avg_view_count.video_view_count.mean()}")
print(f"Median view count of videos with 'claim': {avg_view_count.video_view_count.median()}")
```

```python
# What is the average view count of videos with "opinion" status?
avg_view_count = data[data['claim_status'] == 'opinion']
print(f"Avg. view count of videos with 'opinion': {avg_view_count.video_view_count.mean()}")
print(f"Median view count of videos with 'claim': {avg_view_count.video_view_count.median()}")
```

* The mean and median within each claim category are close to one another, but there is a vast discrepancy between view count for videos labeled as claims and videos labeled as opinons.

Now, examine trends associated with the ban status of the author.

```python
# Get counts for each group combination of claim status and author ban status
# data.groupby(['claim_status', 'author_ban_status']).count()[['#']]
data.groupby(['claim_status', 'author_ban_status']).size()
```

* There's a noticeable difference in the number of banned authors between 'claim' and 'opinion' video. 1439 with authors that are banned. otherwise, 196 have banned authors that are opinion.
    * Claim video are more strictly policed than opnion videos
    * Authors must comply with a stricter set of rules if they post a claim than if they post an opinion.

* The video from banned authors have most significant values compare others.

```python
data.groupby(['author_ban_status']).agg({'video_view_count':['mean','median'],
                                         'video_like_count':['mean','median'],
                                         'video_share_count':['mean', 'median']
                                        })
```

```python
# What's the median video share count of each author ban status?
data.groupby(['author_ban_status']).median(numeric_only=True)[['video_share_count']]
```

* Banned author have a median share count that's 33 times the median share count of active authors! Explore this in more depth.

Group the data by `author_ban_status` with mean, median and count of `video_view_count`, `video_like_count` and `video_share_count` columns.

```python
# authore_bans_status - count, mean, median of each of following columns.
data.groupby(['author_ban_status']).agg({'video_view_count':['count','mean','median'],
                                         'video_like_count':['count','mean','median'],
                                         'video_share_count':['count','mean','median']})
```

* A few observations figure out:
  * Banned authors and those under review get far more views, like, and shares than active authors.
  * Across all ban status, most mean values much higher than median except view count.

```python
# Create a likes_per_view column
data['likes_per_view'] = data['video_like_count']/data['video_view_count']
# Create a comments_per_view column
data['comments_per_view'] = data['video_comment_count'] / data['video_view_count']
# Create a shares_per_view column
data['shares_per_view'] = data['video_share_count']/data['video_view_count']
```

```python
data.groupby(['claim_status', 'author_ban_status']).agg(
    {'likes_per_view':['count','mean','median'],
     'comments_per_view' : ['count','mean','median'],
     'shares_per_view' : ['count','mean','median']
    }
)
```

* We knew that videos by banned and under reviewed got far more views, likes and shares than active authors. However, as a view from engagement rates, the claim status is more related than ban status. 
* Both claim and opinion, most are in an active states however it's notable that in the case of opinion is far more active count and far less than comments.. almost no comments.
